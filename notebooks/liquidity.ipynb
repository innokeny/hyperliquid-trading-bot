```python
import time

def get_candles():
    info = Info(constants.MAINNET_API_URL, skip_ws=True)
    
    # Параметры запроса (пример для ETH)
    params = {
        "type": "candleSnapshot",  # Важно: camelCase!
        "req": {
            "coin": "ETH",         # Торговая пара
            "interval": "1h",      # Доступные интервалы: 15m, 1h, 4h, 12h, 1d
            "startTime": int((time.time() - 86400) * 1000),  # 24 часа назад
            "endTime": int(time.time() * 1000)               # Текущее время
        }
    }
    
    try:
        response = info.candles_snapshot(params)
        print(f"Статус: {response.status_code}")
        print("Свечи:")
        for candle in response:
            print(f"""
            Время: {candle['t']}
            Открытие: {candle['o']}
            Закрытие: {candle['c']}
            Максимум: {candle['h']}
            Минимум: {candle['l']}
            Объем: {candle['v']}
            """)
            
    except Exception as e:
        print(f"Ошибка: {str(e)}")

get_candles()
```
